```python
import pandas as pd
import folium

brasil = folium.Map(
        location=[-16.1237611, -59.9219642],
        zoom_start=4
)
```

```python
brasil
```

```python
empresas = pd.read_csv('empresas.xz')

empresas1 = empresas.iloc[0]
folium.Marker(
    location=[empresas1['latitude'], empresas1['longitude']],
).add_to(brasil)

empresas2 = empresas.iloc[1]
folium.Marker(
    location=[empresas2['latitude'], empresas2['longitude']],
).add_to(brasil)

brasil
```

```python
empresas_pe = empresas[empresas['state']=='PE']
```

```python
pernambuco = folium.Map(
    location=[-8.3833569, -38.5757127],
    zoom_start=7
)
pernambuco
```

```python
for _,emp in empresas_pe.iterrows():
    folium.Marker(
        location=[emp['latitude'], emp['longitude']],
    ).add_to(pernambuco)

pernambuco
```

```python
empresas_olinda_pe = empresas[empresas['city']=='OLINDA']

olinda = folium.Map(
    location=[-7.9981267,-34.9082027],
    zoom_start=13
)

for _,emp in empresas_olinda_pe.iterrows():
    folium.Marker(
        location=[emp['latitude'],emp['longitude']],
    ).add_to(olinda)

olinda
```

```python
colors = {
    'AMPARO':'pink',
    'GUADALUPE':'blue',
    'CASA CAIADA': 'green',
    'PEIXINHOS':'orange',
    'RIO DOCE':'red',
    'BAIRRO NOVO':'purple',
}

olinda = folium.Map(
    location=[-7.9981267, -34.9082027],
    zoom_start=13
)

for _,emp in empresas_olinda_pe.iterrows():
    if emp['neighborhood'] in colors.keys():
        folium.Marker(
            location=[emp['latitude'],emp['longitude']],
            icon=folium.Icon(color=colors[emp['neighborhood']])
        ).add_to(olinda)
```

```python
olinda
```

```python
olinda = folium.Map(
    location=[-7.9981267, -34.9082027],
    zoom_start=13
)

emp = empresas_olinda_pe.iloc[2]
folium.Marker(
    location=[emp['latitude'], emp['longitude']],
    popup='sou uma empresa de pernambuco',
    icon=folium.Icon(color=colors[emp['neighborhood']])
).add_to(olinda)
```
