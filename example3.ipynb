```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
pernambuco = folium.Map(
        location=[-16.1237611, -59.9219642],
        zoom_start=4
)

empresas = pd.read_csv('empresas.xz')
empresas_pe = empresas[empresas['state']=='PE']

marker_cluster = MarkerCluster().add_to(pernambuco)
for _,emp in empresas_pe.iterrows():
    folium.Marker(
        location=[emp['latitude'], emp['longitude']],
    ).add_to(marker_cluster)

pernambuco
```
