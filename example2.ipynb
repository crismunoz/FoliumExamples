```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
pernambuco = folium.Map(
        location=[-16.1237611, -59.9219642],
        zoom_start=4
)

empresas = pd.read_csv('empresas.xz')
empresas_pe = empresas[empresas['state']=='PE']

marker_cluster = MarkerCluster(parse_html=True,).add_to(pernambuco)
for id,emp in empresas_pe.iterrows():
    folium.Marker(
        location=[emp['latitude'], emp['longitude']],
        popup='ID:'+str(id) + ' ' + str(emp.name),
        icon=folium.Icon(color='red', icon_color='white', icon='male', angle=0, prefix='fa'),
    ).add_to(marker_cluster)

# Not showed (Jupyter problem), only in html is ok
pernambuco.save('index.html')
```
